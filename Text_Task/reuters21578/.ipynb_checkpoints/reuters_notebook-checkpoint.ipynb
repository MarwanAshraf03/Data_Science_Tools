```python
import pandas as pd 
```

```python
import re
title_body_dict = {}
for num in range(22):
    file = f"reut2-{num:03d}.sgm"
    with open(file=file, mode="r") as f:
        s = f.read()
    value = re.findall("<BODY>(.*?)</BODY>", s, re.S)
    key = re.findall("<TITLE>(.*)</TITLE>", s)
    for i in range(len(key)):
        title_body_dict[key[i]] = value[i]

all_articles = ""

for i in title_body_dict.keys():
    all_articles += title_body_dict[i]

all_words = all_articles.split()
unique_words = set(all_words)
```

```python
unique_words_list=set(all_words)
```

```python
unique_words_list = list(unique_words)
```

```python
# To filter data =>
# 1) removing this characters from any word "().;,<>?
# 2) remove this sequence &lt from any word
# 3) remove all words less than or equal 2 characters
# 4) remove any word that contains a number
# I don't know why but this loop doesn't filter the words completely in one time it needs to be executed
# for like 2 or 3 times for the @unique_words_list to be completely filtered
# ## TODO: Check why this happens and how to avoid it.

word = 0
while True:
    unique_words_list[word] = unique_words_list[word].lower()
    if len(unique_words_list[word]) <= 2:
        del unique_words_list[word]
        word-=1
        continue
    if "&lt" in unique_words_list[word]:
        unique_words_list[word] = unique_words_list[word].replace("&lt", '')
    for char in unique_words_list[word]:
        if char in "123456789$":
            del unique_words_list[word]
            word-=1
            break
        if char in '"().;,<>?':
            unique_words_list[word] = unique_words_list[word].replace(f'{char}', '')
    word += 1
    if word >= len(unique_words_list):
        break
```

```python
for i in range(len(unique_words_list[0])):
    if i == '"':
        del unique_words_list[0][i]
```

```python
stopwords=[]
with open("stopwords.txt", 'r') as file:

    line = file.readline()

    while line:
        
        stopwords.append(line.strip())  
        
        line = file.readline()
```

```python
stopwords=set(stopwords)
stopwords
```

```python
unique_words_set_filtered=unique_words_set.difference(stopwords)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(vocabulary=unique_words_list_filtered)
```
